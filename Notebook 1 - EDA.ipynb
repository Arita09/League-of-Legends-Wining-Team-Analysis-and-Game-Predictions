We have now merged our data and dropped all the features we didn't need; therefore, let's start the EDA on our merged dataframe by checking if there is any null values.

```python
# Checking for null values
M1_df.isna().sum()
```

---
There are some columns that appear as if they were missing a lot of value like `hexctechs`(type of dragon) or `heralds`, well the truth is they are not really missing any value, simply put: the value for those columns are 0, which means they have not obtained any of those resources in the game. For these type of columns we have decided to place the null value with 0.

Also there are some other columns like `teamname` or `xpat15` that are columns similar to them with such a low variables missing like 2 rows or 22, but provided that it is basically impossible to have no experience(xp15) at 15 minutes, and they are missing such small amount of value, and in that case their relevance in our analysis would be minimal so we have decided to drop them .

---

```python
# Dropping null value rows
M1_df.dropna( how='any',
                          subset=['teamname', 'golddiffat15', 'assistsat15', 'csdiffat15', 'killsat15', 'xpdiffat15','csat15','xpat15','goldat15','earnedgold','earned gpm','firstblood'], inplace = True)
```
